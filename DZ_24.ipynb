Задание 1
Создайте класс Device, который содержит информацию об устройстве.
Спомощью механизма наследования, реализуйте класс CoffeeMachine (содержит информацию о кофемашине), класс Blender (содержит информацию о блендере), класс
MeatGrinder (содержит информацию о мясорубке).
Каждый из классов должен содержать необходимые
для работы методы.

```python
class Device:
    def __init__(self, name, power):
        self.name = name
        self.power = power

    def turn_on(self):
        print(f"{self.name} включен.")

    def turn_off(self):
        print(f"{self.name} выключен.")

    def get_info(self):
        return f"Устройство: {self.name}, Мощность: {self.power} Вт"


class CoffeeMachine(Device):
    def __init__(self, name, power, water_capacity):
        super().__init__(name, power)
        self.water_capacity = water_capacity

    def brew_coffee(self):
        print(f"{self.name} готовит кофе.")

    def get_info(self):
        return f"{super().get_info()}, Вместимость воды: {self.water_capacity} л"


class Blender(Device):
    def __init__(self, name, power, jar_capacity):
        super().__init__(name, power)
        self.jar_capacity = jar_capacity

    def blend(self):
        print(f"{self.name} смешивает ингредиенты.")

    def get_info(self):
        return f"{super().get_info()}, Вместимость чаши: {self.jar_capacity} л"


class MeatGrinder(Device):
    def __init__(self, name, power, meat_capacity):
        super().__init__(name, power)
        self.meat_capacity = meat_capacity

    def grind_meat(self):
        print(f"{self.name} перемалывает мясо.")

    def get_info(self):
        return f"{super().get_info()}, Вместимость мяса: {self.meat_capacity} кг"


coffee_machine = CoffeeMachine("Кофемашина", 1500, 1.5)
blender = Blender("Блендер", 600, 2.0)
meat_grinder = MeatGrinder("Мясорубка", 800, 2.5)

print(coffee_machine.get_info())
coffee_machine.turn_on()
coffee_machine.brew_coffee()
coffee_machine.turn_off()

print(blender.get_info())
blender.turn_on()
blender.blend()
blender.turn_off()

print(meat_grinder.get_info())
meat_grinder.turn_on()
meat_grinder.grind_meat()
meat_grinder.turn_off()
```

Задание 2
Создайте класс Ship, который содержит информацию
о корабле.
С помощью механизма наследования, реализуйте
класс Frigate (содержит информацию о фрегате), класс
Destroyer (содержит информацию об эсминце), класс
Cruiser (содержит информацию о крейсере).
Каждый из классов должен содержать необходимые
для работы методы.

```python
class Ship:
    def __init__(self, name, length, width, speed):
        self.name = name
        self.length = length
        self.width = width
        self.speed = speed

    def sail(self):
        print(f"{self.name} идет на скорости {self.speed} узлов.")

    def get_info(self):
        return f"Корабль: {self.name}, Длина: {self.length} м, Ширина: {self.width} м, Скорость: {self.speed} узлов"


class Frigate(Ship):
    def __init__(self, name, length, width, speed, armament):
        super().__init__(name, length, width, speed)
        self.armament = armament

    def attack(self):
        print(f"{self.name} атакует с помощью {self.armament}.")

    def get_info(self):
        return f"{super().get_info()}, Вооружение: {self.armament}"


class Destroyer(Ship):
    def __init__(self, name, length, width, speed, missile_system):
        super().__init__(name, length, width, speed)
        self.missile_system = missile_system

    def launch_missile(self):
        print(f"{self.name} запускает ракету с помощью {self.missile_system}.")

    def get_info(self):
        return f"{super().get_info()}, Ракетная система: {self.missile_system}"


class Cruiser(Ship):
    def __init__(self, name, length, width, speed, aircraft_capacity):
        super().__init__(name, length, width, speed)
        self.aircraft_capacity = aircraft_capacity

    def launch_aircraft(self):
        print(f"{self.name} поднимает в воздух {self.aircraft_capacity} самолетов.")

    def get_info(self):
        return f"{super().get_info()}, Вместимость для самолетов: {self.aircraft_capacity}"



frigate = Frigate("Фрегат 1", 120, 20, 30, "пушки и ракеты")
destroyer = Destroyer("Эсминец 1", 150, 25, 35, "Гарпун")
cruiser = Cruiser("Крейсер 1", 180, 30, 25, 10)

print(frigate.get_info())
frigate.sail()
frigate.attack()

print(destroyer.get_info())
destroyer.sail()
destroyer.launch_missile()

print(cruiser.get_info())
cruiser.sail()
cruiser.launch_aircraft()
```

Задание 3
Запрограммируйте класс Money (объект класса оперирует одной валютой) для работы с деньгами.
В классе должны быть предусмотрены поле для хранения целой части денег (доллары, евро, гривны и т.д.) и
поле для хранения копеек (центы, евроценты, копейки
и т.д.).
Реализовать методы для вывода суммы на экран, задания значений для частей.

```python
class Money:
    def __init__(self, dollars=0, cents=0):
        self.set_amount(dollars, cents)

    def set_amount(self, dollars, cents):
        """Устанавливает сумму, учитывая правильное преобразование"""
        if cents < 0:
            raise ValueError("Cents cannot be negative.")
        total_cents = dollars * 100 + cents
        if total_cents < 0:
            raise ValueError("Total amount cannot be negative.")
        self.total_cents = total_cents

    def get_dollars(self):
        """Возвращает доллары"""
        return self.total_cents // 100

    def get_cents(self):
        """Возвращает центы"""
        return self.total_cents % 100

    def display(self):
        """Выводит сумму на экран в формате долларов и центов"""
        dollars = self.get_dollars()
        cents = self.get_cents()
        print(f"{dollars} dollars and {cents} cents")

    def add(self, other):
        """Добавляет другую сумму к данной"""
        if not isinstance(other, Money):
            raise TypeError("You can only add Money to Money.")
        self.set_amount(self.get_dollars() + other.get_dollars(), self.get_cents() + other.get_cents())

    def subtract(self, other):
        """Вычитает другую сумму из данной"""
        if not isinstance(other, Money):
            raise TypeError("You can only subtract Money from Money.")
        result_cents = self.total_cents - other.total_cents
        if result_cents < 0:
            raise ValueError("Resulting amount cannot be negative.")
        self.total_cents = result_cents



money1 = Money(10, 50)
money2 = Money(5, 75)


print("Initial amounts:")
money1.display()
money2.display()


money1.add(money2)
print("\nAfter adding:")
money1.display()


money3 = Money(2, 30)
print("\nBefore subtracting:")
money1.display()


money1.subtract(money3)
print("After subtracting:")
money1.display()
```
